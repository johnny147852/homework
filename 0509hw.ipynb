```python
import numpy as np
import scipy.stats as si
```

```python
S, T, r, sigma = 90, 0.5, 0.01, 0.3
#K=[]
#for i in range (1,101):
#    K.append(i)
#print(K[0])
#type(K[0])
```

```python
def BS(S,K,T,r,sigma): 
    d1 = (np.log(S/K)+(r+0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2 = (np.log(S/K)+(r-0.5*sigma**2)*T)/(sigma*np.sqrt(T))    
    call = S*si.norm.cdf(d1,0.0,1.0)-K*np.exp(-r*T)*si.norm.cdf(d2,0.0,1.0)
    put = -S*si.norm.cdf(-d1,0.0,1.0)+K*np.exp(-r*T)*si.norm.cdf(-d2,0.0,1.0)
    return call,put
```

```python
CallPrice=[]
PutPrice=[]
Ks=[]
for K in range(1,150):
    Ks.append(K)
    BS_call, BS_put= BS(S,K,T,r,sigma)
    CallPrice.append(BS_call)
    PutPrice.append(BS_put)
#    print('BS_call=',round(BS_call,4),'BS_put=',round(BS_put,4))
```

```python
import matplotlib.pyplot as plt
plt.plot(Ks,CallPrice,label='call')
plt.plot(Ks,PutPrice,label='put')
plt.legend()
plt.xlabel('Call/Put Price')
plt.ylabel('Exercise Price')
```
